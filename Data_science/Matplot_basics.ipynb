```python
fig=plt.figure(figsize=(8,2))  # IT is 8 inches by 2 Inches

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='Y Cubed')
ax.legend(loc=2)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax.plot(x,x**2,color="Orange")
ax.plot(x,x**2,color="Orange",linewidth=3)  #lw=3 is also for line width #we can also give color code hex code form
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
#ax.plot(x,x**2,color="Orange")
ax.plot(x,x**2,color="Orange",lw=3,alpha=0.5) #Transparency in line
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
#ax.plot(x,x**2,color="Orange")
ax.plot(x,x**2,color="Orange",lw=3,alpha=0.5,linestyle="--") #Line Style
#ax.plot(x,x**2,color="Orange",lw=3,alpha=0.5,linestyle=":") #Line Style
```
