```python
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
file = "file.wav"
```

```python
signal,sample_rate = librosa.load(file, sr=44100)
```

```python
librosa.display.waveplot(signal, sr =sample_rate)
```

```python
fft = np.fft.fft(signal)
print('lenght of fft :',len(fft),'\n',fft)
```

```python
magnitude = np.abs(fft)
```

```python
frequency = np.linspace(0 ,sample_rate ,len(magnitude))
```

```python
plt.plot(frequency,magnitude)
plt.xlabel("Frequency")
plt.ylabel("magnitude")
```

```python
left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(magnitude)/2)]
```

```python
plt.plot(left_frequency,left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("magnitude")
```

```python
n_fft = 2048
hop_lenght = 512
```

```python
stft = librosa.core.stft(signal, hop_length=hop_lenght, n_fft= n_fft)
spectrogram = np.abs(stft)
```

```python
librosa.display.specshow(spectrogram,sr = sample_rate ,hop_length=hop_lenght)
plt.colorbar()
plt.ylabel("Frequency")
plt.xlabel("time")
```

```python
log_spectrogram = librosa.amplitude_to_db(spectrogram)
```

```python
librosa.display.specshow(log_spectrogram,sr = sample_rate,hop_length=hop_lenght)
plt.colorbar()
```

```python
MFFCs = librosa.feature.mfcc(signal,n_fft=n_fft, n_mfcc=12)
```

```python
librosa.display.specshow(MFFCs , sr=sample_rate,hop_length=hop_lenght)
plt.xlabel("Time")
plt.ylabel("MFCC")
```
